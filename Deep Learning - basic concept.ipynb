```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

from keras.models import Sequential
from keras.layers import Dense
```

```python
pima_diabetes = pd.read_csv('D:\Kuliah\Course Online\Mentoring Deep Learning\File code\data\pima_diabetes.csv', names=['preg_times','glukose','blood_pressure','skin_thickness','insulin','BMI','pedigree','age','class'])
```

## Split datasets

```python
features = pima_diabetes.values[:,:8]
features.shape
```

```python
labels = pima_diabetes.values[:,8:]
labels.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=20)
```

## define model

```python
model = Sequential()
```

```python
model.add(Dense(16, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

# Compile model

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Train model

```python
model.fit(features,labels, epochs=150, batch_size=12)
```

```python
plt.figure(figsize=[12,6])
plt.plot(model.history.history['acc'], 'r')
plt.legend(['Training acc'])
plt.xlabel('Epochs',fontsize=12)
plt.ylabel('Accuracy',fontsize=12)
plt.title('Accuracy curves', fontsize=16)

plt.show()
```

## evaluate model

```python
_, accuracy = model.evaluate(X_test,y_test, verbose=False)
```

```python
print(f'Accuracy:{accuracy*100:.2f}')
```
